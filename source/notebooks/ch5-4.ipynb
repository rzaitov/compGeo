```python
# Import libraries
import numpy as np
import matplotlib.pyplot as plt

# Import function DownPlunge
import sys, os
sys.path.append(os.path.abspath('../functions'))

# Old API
from DownPlunge import DownPlunge

# New API
from down_plunge import down_plunge
```

```python
# Read the tops from the text files
jtc = np.loadtxt(os.path.abspath('../data/ch5-4/jtc.txt'))
js = np.loadtxt(os.path.abspath('../data/ch5-4/js.txt'))
kp = np.loadtxt(os.path.abspath('../data/ch5-4/kp.txt'))
```

```python
# old api: DownPlunge usage.
# Trend and plunge of the fold axis in radians
trend = 125 * np.pi/180
plunge = 26 * np.pi/180

# Transform the points
jtcdp = DownPlunge(jtc,trend,plunge)
jsdp = DownPlunge (js,trend,plunge)
kpdp = DownPlunge(kp,trend,plunge)

# Plot the down plunge section
plt.plot(jtcdp[:,1],jtcdp[:,0],'k-')
plt.plot(jsdp[:,1],jsdp[:,0],'r-')
plt.plot(kpdp[:,1],kpdp[:,0],'b-')

# Display the section's orientation
# Notice that the fold axis plunges SE
# Therefore the left side of the section is NE
# and the right side is SW
plt.text(-2.1e4,12e3,'NE')
plt.text(-0.6e4,12e3,'SW')

# Make axes equal
plt.axis('equal');
```

```python
# new api with down_plunge.
# Trend and plunge of the fold axis in radians
trend, plunge = np.radians([ 125, 26 ])

T = down_plunge(trend, plunge)
# Transform the points.
jtc_dp = np.dot(T, jtc.T)
js_dp = np.dot(T, js.T)
kp_dp = np.dot(T, kp.T)

# Plot the down plunge section
plt.plot(jtc_dp[1, :], jtc_dp[0, :], 'k-')
plt.plot(js_dp[1, :], js_dp[0, :], 'r-')
plt.plot(kp_dp[1, :], kp_dp[0, :], 'b-')

# Display the section's orientation
# Notice that the fold axis plunges SE
# Therefore the left side of the section is NE
# and the right side is SW
plt.text(-2.1e4, 12e3, 'NE')
plt.text(-0.6e4, 12e3, 'SW')

# Make axes equal
plt.axis('equal');
```
